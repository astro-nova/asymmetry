# HST vs SDSS

#### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy.io import fits
from statmorph import source_morphology
from astropy.visualization import AsinhStretch, AsymmetricPercentileInterval
from astropy.wcs import WCS
from photutils.segmentation import detect_sources
from scipy.stats import nobrm as gauss
from skimage import transform as T
norm = AsymmetricPercentileInterval(1, 99.5)
stretch = AsinhStretch(0.2)
```

## Load HST data

```python
ra = 181.49904
dec = -2.90890
```

Load in the FITS files:

```python
hst_file = fits.open(f'../data/hst/data/data/F438W/E12.fits') 
```

```python
pxscale = hst_file[0].header['D001SCAL']
wcs = WCS(hst_file[1].header)

yc, xc = wcs.world_to_pixel_values(np.array([[ra, dec]]))[0]
yc, xc = int(xc), int(yc)
r = int(10/pxscale)

hst_img = hst_file[1].data[yc-r:yc+r, xc-r:xc+r]
```

```python
plt.imshow(hst_img, vmin=0, vmax=0.1)
```

## Load SDSS data

```python
sdss_file = fits.open(f'../data/sdss/E12.fits') 
```

```python
pxscale = 0.396
wcs = WCS(sdss_file[0].header)

yc, xc = wcs.world_to_pixel_values(np.array([[ra, dec]]))[0]
yc, xc = int(xc), int(yc)
r = int(10/pxscale)

sdss_img = sdss_file[0].data[yc-r:yc+r, xc-r:xc+r]
```

## HST morphology

```python
hst_sigma = np.std(hst_img[:50, :50])
hst_mean = np.mean(hst_img[:50, :50])
hst_img = hst_img - hst_mean

segmap_mask = detect_sources(hst_img, 5*hst_sigma, npixels=5)
segmap_mask = segmap_mask.data
segmap_mask[100:250, 100:250] = 0
segmap_mask[segmap_mask > 0] = 1

segmap_mask2 = detect_sources(hst_img, 1*hst_sigma, npixels=200)
segmap_mask2 = segmap_mask2.data
segmap_mask2[100:280, 110:300] = 0

segmap_mask = segmap_mask + segmap_mask2
segmap_mask = segmap_mask > 0

fill_vals = gauss.rvs(loc=0, scale=hst_sigma, size=hst_img.shape)
arr = np.ma.array(hst_img, mask=segmap_mask)
arr = arr.filled(fill_vals)
hst_img = np.ma.getdata(arr)

xc, yc = int(hst_img.shape[1]/2), int(hst_img.shape[0]/2)
```

Calculate asymmetry

```python
hst_img_rotated = T.rotate(hst_img, 180, center=np.array([xc, yc]), order=0)

N = hst_img.shape[0]
bg = hst_img[:50,:50]
bg_a = np.sum(np.abs(bg - bg[::-1])) / 2500

top = np.sum(np.abs(hst_img-hst_img_rotated)) - bg_a*N**2
bot = np.sum(np.abs(hst_img))
a_cas_hst = top/bot

print(a_cas_hst)
```

```python
top = np.sum((hst_img-hst_img_rotated)**2) - 2*hst_sigma**2
bot = np.sum(hst_img**2) - hst_sigma**2
a_sq_hst = (top/bot)
a_sq_hst
```

## SDSS morphology

```python
sdss_sigma = np.std(sdss_img[:10, :10])
sdss_mean = np.mean(hst_img[:10, :10])
sdss_img = sdss_img - sdss_mean

segmap_mask = detect_sources(sdss_img, 17*hst_sigma, npixels=10)
segmap_mask = segmap_mask.data
segmap_mask[segmap_mask == 1] = 0
segmap_mask[segmap_mask > 0] = 1


fill_vals = gauss.rvs(loc=0, scale=sdss_sigma, size=sdss_img.shape)
arr = np.ma.array(sdss_img, mask=segmap_mask)
arr = arr.filled(fill_vals)
sdss_img = np.ma.getdata(arr)

xc_sdss, yc_sdss = int(sdss_img.shape[1]/2), int(sdss_img.shape[0]/2)
```

```python
sdss_img_rotated = T.rotate(sdss_img, 180, center=np.array([xc_sdss, yc_sdss]), order=0)

N = sdss_img.shape[0]
bg = sdss_img[:10,:10]
bg_a = np.sum(np.abs(bg - bg[::-1])) / 100

top = np.sum(np.abs(sdss_img-sdss_img_rotated)) - bg_a*N**2
bot = np.sum(np.abs(sdss_img))
a_cas_sdss = top/bot

print(a_cas_sdss)
```

```python
top = np.sum((sdss_img-sdss_img_rotated)**2) - 2*sdss_sigma**2
bot = np.sum(sdss_img**2) - sdss_sigma**2
a_sq_sdss = (top/bot)
a_sq_sdss
```

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
rvs = stats.norm.rvs(loc=0, scale=1, size=(5000, 10000))
```

```python
rvs[:,0][np.newaxis,:].shape
```

```python
randints = np.random.randint(0, 10000, size=10)
bs_var_arrs = []
for i in randints:
    variances_bs = stats.bootstrap(rvs[:,i][np.newaxis,:], np.var, n_resamples=1000)
    bs_var_arrs.append(variances_bs.bootstrap_distribution)
```

```python
variances_sim = np.var(rvs, axis=1)
```

```python
bins = np.linspace(0.9,1.1,51)
plt.hist(variances_sim, bins=bins, density=True, alpha=1, label='10k samples');
for arr in bs_var_arrs:
    plt.hist(arr, bins=bins, density=True, alpha=0.1, label='10k bootstraps');
# plt.legend()
```
